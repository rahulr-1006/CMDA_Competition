```python
selected_conditions = [
    "Coronary heart disease",
    "Current cigarette smoking",
    "COPD, emphysema, chronic bronchitis",  # Corrected typo
    "Diagnosed diabetes, self-reported",
    "Obesity, self-reported",
    "Asthma episode/attack",
    "Hypertension diagnosis, self-reported",
    "Cervical cancer"
]

# Normalize case and remove leading/trailing spaces for robust filtering
health_conditions_filtered = health_conditions[
    health_conditions["TOPIC"].str.strip().str.lower().isin(
        [condition.strip().lower() for condition in selected_conditions]
    )
]

print("Unique Topics in the filtered dataset:")
print(health_conditions_filtered["TOPIC"].unique())
```

```python
# Drop unnecessary columns
columns_to_drop = [
    "SUBTOPIC", "SUBTOPIC_ID", "STANDARD_ERROR", 
    "ESTIMATE_LCI", "ESTIMATE_UCI", "TIME_PERIOD_ID"
]
health_conditions_filtered = health_conditions_filtered.drop(columns=columns_to_drop)

# Rename columns for better clarity
health_conditions_filtered.rename(
    columns={
        "TOPIC": "condition",
        "SUBGROUP": "demographic_group",
        "TIME_PERIOD": "year",
        "ESTIMATE": "value"
    },
    inplace=True
)

# Map demographic groups to simpler categories
group_mapping = {
    "18 years and older": "all_ages",
    "Black only": "black",
    "White only": "white",
    "Asian only": "asian",
    "Hispanic": "hispanic",
    "Non-Hispanic": "non_hispanic",
    "Female": "female",
}

health_conditions_filtered["demographic_group"] = (
    health_conditions_filtered["demographic_group"].map(group_mapping).fillna("other")
)
```

```python
# Create unique condition + demographic group identifier
health_conditions_filtered["condition_group"] = (
    health_conditions_filtered["condition"] + "_" + health_conditions_filtered["demographic_group"]
).str.lower().str.replace(" ", "_").str.replace(",", "_")

# Pivot the dataset
health_conditions_pivot = health_conditions_filtered.pivot_table(
    index="year",  # Rows are years
    columns="condition_group",  # Columns are condition + demographic group
    values="value",  # Values are the estimate
    aggfunc="mean"  # Handle any duplicates by averaging
).reset_index()

# Flatten multi-level column names
health_conditions_pivot.columns = ["year"] + [
    col.lower().replace(" ", "_") for col in health_conditions_pivot.columns[1:]
]
```

```python
# Define exceptions (conditions that inherently apply to females only)
exceptions = ["cervical_cancer"]

# Define demographic subgroups to scale
subgroups = ["asian", "black", "white", "other"]

# Iterate through all columns
for col in health_conditions_pivot.columns:
    if "_all_ages" in col and not any(exception in col for exception in exceptions):
        # Get the base condition name
        base_condition = col.replace("_all_ages", "")
        
        # Find the corresponding female column
        female_col = f"{base_condition}_female"
        
        if female_col in health_conditions_pivot.columns:
            # Calculate scaling factor
            scaling_factor = (
                health_conditions_pivot[female_col] / health_conditions_pivot[col]
            )
            
            # Scale all demographic subgroup columns
            for subgroup in subgroups:
                subgroup_col = f"{base_condition}_{subgroup}"
                if subgroup_col in health_conditions_pivot.columns:
                    health_conditions_pivot[subgroup_col] *= scaling_factor

# Verify the results
display(health_conditions_pivot)

## FINAL
```

```python
import pandas as pd
import glob

# Define the file paths for all Excel files
file_paths = [
    "/Users/rahulramakrishnan/Downloads/womens-earnings-tables-2018.xlsx",
    "/Users/rahulramakrishnan/Downloads/womens-earnings-tables-2019 (1).xlsx",
    "/Users/rahulramakrishnan/Downloads/womens-earnings-tables-2020.xlsx",
    "/Users/rahulramakrishnan/Downloads/womens-earnings-tables-2021.xlsx",
    "/Users/rahulramakrishnan/Downloads/womens-earnings-tables-2022.xlsx",
    "/Users/rahulramakrishnan/Downloads/womens-earnings-tables-2023.xlsx",
]

# Initialize an empty list to hold the cleaned DataFrames
cleaned_dataframes = []

# Function to clean a single file
def clean_file(file_path):
    print(f"Processing file: {file_path}")
    # Load the specific sheet with Table 16
    df = pd.read_excel(file_path, sheet_name="Table 18", header=4)

    # Drop any completely empty rows
    df.dropna(how="all", inplace=True)

    # Rename the columns
    df.columns = ['Year', 'Total, 16 years and older', 'White', 'Black', 'Asian', 'Hispanic']

    # Drop rows where 'Year' is NaN or not numeric
    df = df[df['Year'].apply(lambda x: str(x).isdigit())]

    # Convert 'Year' column to integer
    df['Year'] = df['Year'].astype(int)

    # Return the cleaned DataFrame
    return df

# Process each file and append the cleaned DataFrame to the list
for file in file_paths:
    try:
        cleaned_df = clean_file(file)
        cleaned_dataframes.append(cleaned_df)
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Combine all the cleaned DataFrames into a single DataFrame
combined_data = pd.concat(cleaned_dataframes, ignore_index=True)

week_wage = combined_data

# Rename columns in the weekly wages dataset
week_wage.columns = [
    'year', 
    'total_weekly_wage', 
    'white_weekly_wage', 
    'black_weekly_wage', 
    'asian_weekly_wage', 
    'hispanic_weekly_wage'
]

# Convert relevant columns to numeric to ensure consistency
for col in week_wage.columns[1:]:
    week_wage[col] = pd.to_numeric(week_wage[col], errors='coerce')

# Display the cleaned weekly wages dataset
display(week_wage)
```

```python
import pandas as pd
import numpy as np

# Define the target range
TARGET_YEARS = range(2017, 2025)

# Function to check and standardize the 'year' column
def ensure_year_column(df, name):
    if "Year" in df.columns:
        df = df.rename(columns={"Year": "year"})
    if "year" not in df.columns:
        raise KeyError(f"'year' column is missing in the dataset: {name}")
    return df

# Function to clean non-numeric data
def clean_numeric_columns(df):
    for col in df.select_dtypes(include=["object"]).columns:
        # Replace non-numeric values with NaN
        df[col] = pd.to_numeric(df[col], errors="coerce")
    return df

# Function to interpolate datasets to the target year range
def interpolate_to_target_years(df, name):
    print(f"Interpolating {name} to {min(TARGET_YEARS)}-{max(TARGET_YEARS)}")
    # Ensure 'year' column exists and is unique
    df = ensure_year_column(df, name)
    df.set_index("year", inplace=True)

    # Clean non-numeric data
    df = clean_numeric_columns(df)

    # Remove duplicate years by averaging duplicate entries
    df = df.groupby(level=0).mean()

    # Reindex to the target years
    df = df.reindex(TARGET_YEARS)

    # Interpolate missing values (linear)
    df.interpolate(method="linear", inplace=True)

    # Reset index to keep 'year' as a column
    df.reset_index(inplace=True)
    return df

# List of datasets and their names
datasets = [
    infant_mortality_cleaned,
    health_conditions_pivot,
    week_wage,
    maternal_deaths,
    reshaped_college_data,
]

dataset_names = ["infant mortality", "health conditions", "weekly wage", "maternal deaths", "college attainment"]

# Step 1: Interpolate each dataset
interpolated_datasets = []
for dataset, name in zip(datasets, dataset_names):
    interpolated_datasets.append(interpolate_to_target_years(dataset, name))

# Step 2: Merge datasets on 'year'
merged_data = interpolated_datasets[0]
for dataset in interpolated_datasets[1:]:
    merged_data = merged_data.merge(dataset, on="year", how="outer")

# Step 3: Preview the merged dataset
print("Final Merged Data:")
display(merged_data)

# List of columns to drop
columns_to_drop = [
    "Data As Of",
    "Jurisdiction",
    "Group",
    "Subgroup",
    "Month of Death",
    "Time Period",
    "Month Ending Date",
    "Footnote"
]

# Drop the columns from the DataFrame
cleaned_df = merged_data.drop(columns=columns_to_drop, errors='ignore')







# Save the final merged dataset to a CSV file
```

```python
import numpy as np
from scipy.stats import zscore

# Fill missing values with randomization
def fill_missing_values_randomized(df):
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:  # Numeric columns
            median = df[col].median()
            std_dev = df[col].std() * 0.1  # Small fraction of the column's standard deviation
            df[col] = df[col].apply(lambda x: median + np.random.normal(0, std_dev) if np.isnan(x) else x)
        else:  # Non-numeric columns
            mode_values = df[col].mode()
            if not mode_values.empty:
                mode = mode_values[0]
                df[col] = df[col].apply(lambda x: np.random.choice(mode_values) if pd.isnull(x) else x)
    return df

# Apply the function
merged_data = fill_missing_values_randomized(merged_data)
```
